```python
import pandas as pd
import os

# On construit le chemin absolu automatiquement
base_dir = os.path.expanduser("~/work/Chess.com/data")
file_path = os.path.join(base_dir, "stat_parties.parquet")

if os.path.exists(file_path):
    df = pd.read_parquet(file_path)
    print("✅ Fichier chargé avec succès !\n")
    
    # --- RÉSUMÉ DES STATS ---
    print("🏆 NOMBRE DE JOUEURS PAR CLASSE :")
    print(df['class'].value_counts())
    
    print("\n⏱️ TEMPS MOYEN D'OUVERTURE PAR NIVEAU :")
    # On vérifie si les maîtres sont effectivement plus rapides
    stats_temps = df.groupby('class')['avg_opening_move_time'].mean().sort_values()
    print(stats_temps)
    
    print("\n♟️ EXEMPLE DE RÉPERTOIRE (Top 5 Win) :")
    print(df[['username', 'class', 'top_5_openings_win']].head(10))
else:
    print(f"❌ Python ne voit toujours rien dans : {file_path}")
    print("Contenu réel du dossier data :", os.listdir(base_dir) if os.path.exists(base_dir) else "Dossier data introuvable")

def apercu_stats_parquet(file_path):
    # 1. Chargement du fichier
    df = pd.read_parquet(file_path)
    
    print("=== 📊 RÉSUMÉ GÉNÉRAL DU DATASET ===")
    print(f"Nombre de joueurs (lignes) : {df.shape[0]}")
    print(f"Nombre de variables (colonnes) : {df.shape[1]}")
    print("-" * 40)
    
    # 2. Statistiques sur les colonnes numériques (Rating, Temps moyen, etc.)
    print("\n=== 📈 STATS NUMÉRIQUES ===")
    # .T permet de transposer pour mieux lire si tu as beaucoup de colonnes
    print(df.describe().T[['mean', 'std', 'min', '50%', 'max']])
    
    # 3. Répartition par niveau (Class)
    if 'class' in df.columns:
        print("\n=== 🏆 RÉPARTITION PAR CLASSE ===")
        print(df['class'].value_counts())
    
    # 4. Aperçu des types de données et valeurs manquantes
    print("\n=== 🔍 INFOS STRUCTURE ET MANQUANTS ===")
    info_df = pd.DataFrame({
        'Type': df.dtypes,
        'Valeurs Manquantes': df.isnull().sum(),
        '% Manquant': (df.isnull().sum() / len(df) * 100).round(2)
    })
    print(info_df)

    # 5. Top 3 des ouvertures les plus récurrentes dans le dataset (si la colonne existe)
    print("\n=== ♟️ OUVERTURES LES PLUS POPULAIRES (Global) ===")
    # On prend la première ouverture citée dans la liste "top_5"
    first_opening = df['top_5_openings_used'].str.split(',').str[0]
    print(first_opening.value_counts().head(5))
```

```python
apercu_stats_parquet('../data/stat_parties.parquet')
```

```python
def liste_colonnes_parquet(file_path):
    df = pd.read_parquet(file_path)
    print(f"--- Colonnes de {file_path} ---")
    print(df.columns.tolist())
    print("\n--- Types de données ---")
    print(df.dtypes)

# Utilisation
liste_colonnes_parquet('../data/stat_parties.parquet')
```
